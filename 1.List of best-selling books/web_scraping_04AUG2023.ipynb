# Web Scrapping from Wikipedia

### List of best-selling books


We are to create a local dataset containing the table of _List of best-selling books_ from Wikipedia's website:

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_best-selling_books'
page = requests.get(url).text
```

```python
soup = BeautifulSoup(page,
                     'html.parser')
#print(soup)
# output is too massive
```

```python
#print(soup.prettify())
# output is too massive
```

```python
soup.find('table')
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[2]
print(table)
```

```python
html_headers = table.find_all('th')
print (html_headers)
```

```python
raw_headers = [title.text.strip() for title in html_headers]

print(raw_headers)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = raw_headers)
```

```python
column_data = table.find_all('tr')
column_data
#tabel row element
```

```python
len(column_data)
```

```python
column_data[-1]
#we confirm this is the table we want to fetch ( spotted 'The Naked Ape')
```

```python
column_data[0]
# and we want to exclude the headers already fetched them 
```

```python
length = len(df)
length
```

```python
for row in column_data[1:]:
    html_data = row.find_all('td')
    print(html_data)
    cell_row_data = [data.text.strip() for data in html_data]
   
    
    length = len(df)
    df.loc[length] = cell_row_data
```
